```python
import pandas as pd
```

```python
def val_count(x):
    x = x.query('Категория != "Пополнения"')
    x['Сумма платежа'] = abs(x['Сумма платежа'])
    x = x.pivot_table(index='Категория', values='Сумма платежа', aggfunc='sum').sort_values(by='Сумма платежа',
                                                                                      ascending=False)
    return x
    
```
